```python
var1["C"] = var1["A"] +var1["B"]
var1
```

```python
var1["C"] = var1["A"]- var1["B"]
var1
```

```python
var1["C"] = var1["A"] / var1["B"]
var1
```

```python
var1["C"] = var1["A"] // var1["B"]
var1
```
